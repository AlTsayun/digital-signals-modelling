```python
import common
import importlib
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import widgets
importlib.reload(common)
from common import *

def output(strA,  strF, strPhi, maxN):
    try:
        step = 1 / float(maxN)
        xs = np.arange(0, 4, 0.01)
        for a in parseArray(strA):
            for f in parseArray(strF):
                for phi in parseArray(strPhi):
                    func = getSampledFunc(getSawFunc(a, f, phi), step)
                    plt.plot(xs, [func(x) for x in xs])
        plt.rcParams['figure.figsize'] = [16, 5]
        plt.show()
    except ValueError:
        print("Illegal values")

widgets.interact(output, 
                 strA=widgets.Text(description="A", value="x"), 
                 strF=widgets.Text(description="f", value="1, 1 / (0.1 + x)"),
                 strPhi=widgets.Text(description="Phi", value="0"),
                 maxN=widgets.SelectionSlider(options=getPowersOfTwo(0, 15), value=32, description="N"))
```
